```python
data_copy = cleaned_data
data_copy
```

```python
data_copy = pd.get_dummies(cleaned_data, columns=["homeworld", "unit_type"], dtype=int)
data_copy
```

```python
homeworlds = data_copy.loc[:, data_copy.columns.str.startswith('homeworld')]
homeworlds
```

```python
unit_type = data_copy.loc[:, data_copy.columns.str.startswith('unit_type')]
unit_type
```

```python
unit_type.reset_index(drop=True, inplace=True)
homeworlds.reset_index(drop=True, inplace=True)

new_data = pd.concat([homeworlds, unit_type], axis=1)

new_data
```

```python
predicted = trained_model.predict(new_data)
```

```python
cleaned_data['Predictions'] = predicted
cleaned_data
```
